```python
import math as m
```

```python
def f(x):
    return (2*x+1)**2-5*m.cos(10*x)
```

```python
def aurea(x1,x2):
    b = (m.sqrt(5)-1)/2
    a = b*b
    x3 = 0
    x4 = 0
    for i in range(30): 
        print("%i || X1 = %.5f || X2 = %.5f || X3 = %.5f || X4 = %.5f" %(i,x1,x2,x3,x4))
        x3 = x1 + a*(x2-x1)
        x4 = x1 + b*(x2-x1)
        if f(x3) < f(x4):
            x2 = x4
            x4 = x3 
        else: 
            x1 = x3 
            x3 = x4 
    print("%i || X1 = %.5f || X2 = %.5f || X3 = %.5f || X4 = %.5f" %(i,x1,x2,x3,x4))
    return [x1,x2,x3,x4]
```

```python
aurea(-1,0)
```
