```python
import math as m
```

```python
def f(x,y):
    return y*y - 2*x*y - 6*y + 2*x*x + 12 
def dfx(x,y):
    return -2*y + 4*x
def dfy(x,y):
    return 2*y - 2*x - 6 
```

```python
def gradient(x,y,h):
    for i in range(60):
        print("%i || X = %.5f || Y = %.5f" %(i,x,y))
        xn = x - h*dfx(x,y)
        yn = y - h*dfy(x,y)
        if f(xn,yn)< f(x,y):
            h*=2
            y = yn 
            x = xn 
        else: 
            h/=2
        print("%i || X = %.5f || Y = %.5f" %(i,x,y))
```

```python
gradient(1,1,1)
```
